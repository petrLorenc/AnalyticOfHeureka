```python
import json
from pprint import pprint
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC as SVM

from sklearn.model_selection import cross_val_score



treshold = 88

with open('./json/file_preprocess.json') as data_file:
    data = json.load(data_file)

words = []
label = []

print (len(data))
plus = 0

for x in range(len(data)):
    text = ""
    for y in range(len(data[x]["review"])):
        text += data[x]["review"][y] + " "
    words.append(text)
    try:
        if (int(data[x]["rating"][:-1]) > treshold):
            label.append(1)
            plus += 1;
        else:
            label.append(0)
    except ValueError:
        label.append(0)

print(plus)
print(str(plus) + "/" + str(len(data)) + "=" + str(plus/float(len(data))))
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()

tfidf_words = vectorizer.fit_transform(words)
clf = MultinomialNB()
clf2 = SVM()

# clf3 = SVM(kernel='linear', C=1)
# scores = cross_val_score(clf3, tfidf_words, label, cv=10)
# print (scores)                                              

from sklearn.metrics import roc_auc_score

X_train, X_test, y_train, y_test = train_test_split(tfidf_words, label, test_size=0.4,random_state=2)

clf.fit(X_train, y_train)
clf2.fit(X_train, y_train)

print (str(clf.score(X_test, y_test)))
print (str(clf2.score(X_test, y_test)))
```

```python
y_score = clf2.fit(X_train, y_train).decision_function(X_test)
roc_auc_score(y_test, y_score)
```
